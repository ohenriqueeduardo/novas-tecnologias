## Tabela de frequência e um gráfico para "survival"

### Tabela de Frequência

```python
print(f"{df['PassengerId'].value_counts()}\n")
print(f"{df['Survived'].value_counts()}\n")
print(f"{df['Pclass'].value_counts()}\n")
print(f"{df['Name'].value_counts()}\n")
print(f"{df['Sex'].value_counts()}\n")
print(f"{df['Age'].value_counts()}\n")
print(f"{df['SibSp'].value_counts()}\n")
print(f"{df['Parch'].value_counts()}\n")
print(f"{df['Ticket'].value_counts()}\n")
print(f"{df['Fare'].value_counts()}\n")
print(f"{df['Cabin'].value_counts()}\n")
print(f"{df['Embarked'].value_counts()}\n")
```

### Gráfico de Survival

Eu escolhi o gráfico de pizza, o pie, para representar a porcentagem de sobreviventes, justo por demonstrar uma parte mais exata da porcentagem de sobreviventes ou não

```python
survival = df['Survived'].value_counts()

plt.pie(survival.values, labels=["0: Mortos", "1: Sobreviventes"], autopct="%1.1f%%")

plt.title("Contagem de Sobreviventes")
plt.show()
```

## Avalie a relação entre a variável "sex" e a variável "survival". Crie uma tabela de frequência e um gráfico. Mulheres sobreviveram mais que homes?

Tabela de Frequência comparativa

```python
tabela_comp = pd.crosstab(df['Survived'], df['Sex'])

tabela_comp.index = ["Mulheres", "Homens"]
tabela_comp.columns = ["Sobreviventes", "Mortos"]

print(tabela_comp)
```

Gráfico comparativo

```python
tabela_comp.plot(kind='barh')
```

Mulheres sobreviveram mais que homens?

- A resposta é clara, não, elas não sobreviveram mais que os homens, é nítido a diferença entre os sobreviventes homens e as sobreviventes mulheres

## Qual é  distribuição das idades dos passageiros? Crie um gráfico para visualizar a distribuição. Parece simétrica? A maioria dos passageiros era jovem ou idosa?

### Distribuição da idade dos passageiros

```python
plt.hist(df['Age'], bins=20)

plt.title("Distribuição da idade dos passageiros")
plt.xlabel("Idade")
plt.ylabel("Quantidade de passageiross")
plt.show()
```

- É notavel que o gráfico não tem uma simetria tão grande, com um acúmulo maior na parte de 20, 30 anos, o que demonstra também que a população de passageiros era mais jovem que idosa
