## Separa a base em 3 de acordo ao ano de registro:
- Menor que 1975
- Maior ou igual a 1975 e menor que 2000
- Maior ou igual a 2000

```python
dados_p1 = dados[dados["Ano "] < 1975]
dados_p2 = dados[((dados["Ano "]>= 1975) & (dados["Ano "] < 2000))]
dados_p3 = dados[dados["Ano "] >= 2000]
```
