### Desafio - Verificar números primos

Crie um código que receba um número N do usuário e imprima uma mensagem "Número primo" ou "número não=-primo" conforme o teste de primalidade de N

#### Um número é primo quando é divísivel somente por 1 e por ele mesmo.

Verifique o seu algoritmo com os seguintes casos: 

- 2 = True
- 9 = False
- 23 = True
- 49 = False
- 100 = False
- 10007 = True
- 9999991 = True
- 1000000000039 = True
- 170141183460469231731687303715884105727 = True





```python
import math

num = int(input("Insira o número a ser analisado: "))
raiz = math.sqrt(num)

eh_primo = True
if num <= 1:
   eh_primo = False
elif num % 2 == 0 and num > 2:
   eh_primo = False

for i in range(3, int(raiz) + 1, 2):
   if num % i == 0:
      eh_primo = False
      break

if eh_primo:
    print(f"{num} = True.")
else:
    print(f"{num} = False.")
```

```python
carro1 = {
   "modelo": "Skyline R34",
   "ano": 1999,
   "preco": 15000000
}

carro2 = {
   "modelo": "Dodge Charger R/T",
   "ano": 1970,
   "preco": 23000000
}

print(carro1)

#Imprime um valor específico
print(carro2["modelo"])
```

### Função

```python
def calcular_parcela(valor_carro, qtd_parcelas, taxa):
   return ((valor_carro + (valor_carro * taxa)) / qtd_parcelas)

taxa_financiamento = 1.05
qtd_parcelas = 48

valor_parcela_carro1 = calcular_parcela(carro1["preco"], qtd_parcelas, taxa_financiamento)
valor_parcela_carro2 = calcular_parcela(carro2["preco"], qtd_parcelas, taxa_financiamento)

print(f"O valor da parcela do {carro1["modelo"]} é de: R${valor_parcela_carro1:.2f}")
print(f"O valor da parcela do {carro2["modelo"]} é de: R${valor_parcela_carro2:.2f}")
```
