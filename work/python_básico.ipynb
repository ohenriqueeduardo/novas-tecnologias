### Operadores Aritméticos

```python
A = 2
B = 3

soma = A + B
sub = A - B
mult = A * B
div = A / B
poten = A ** B
modulo = A % B

print(soma)
print(sub)
print(mult)
print(poten)
print(modulo)
```

### Operadores Lógicos

```python
A = 2
B = 3

print(A > B)
print(A < B)
print(A == B)
print(A != B)
```

### Estruturas Condicionais

```python
idade = int(input("Insira a sua idade: "))
possui_habilitacao = True

if idade >= 18 & possui_habilitacao:
    print("\nPossui habilitação e é maior de idade")
elif idade >= 18:
    print("\nMaior de idade")
else:
    print("\nMenor de idade")
```

### Estruturas de Repetição

```python
for i in range(5):
    print(i)

cont = 0
print("-" * 30)
while cont < 5:
    print(cont)
    cont += 1
```
